1. Evaluate Logistic Regression, SVM (linear Kernel), and SVM (gaussian kernel) on sklearn breast cancer dataset.

2. For each model, (a) plot recall vs number of features and (b) precision vs number of features

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
# Making a function to solve each problem

def problemI(model):
  y_pred = model.fit(x_train, y_train).predict(x_test)
  print("Precision Score:", precision_score(y_test, y_pred))
  print("Recall Score:", recall_score(y_test, y_pred))
  print("\nConfusion Matrix:")
  ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=cancer.target_names).plot()
  plt.show()

def problemII(model):
  recalls = []
  precision = []
  for i in range(1, 31):
      y_pred = model.fit(x_train[:, :i], y_train).predict(x_test[:, :i])
      recalls.append(recall_score(y_test, y_pred))
      precision.append(precision_score(y_test, y_pred))
  # (a) Recall vs #features
  plt.plot(range(1, 31), recalls)
  plt.xlabel("# of features")
  plt.ylabel("Recall")
  plt.title("Recall vs # of features")
  plt.show()

  # (b) Precision vs #features curves.
  plt.plot(range(1, 31), precision)
  plt.xlabel(" # of features")
  plt.ylabel("Precision")
  plt.title("Precision vs # of features")
  plt.show()
```

```python
#split the data into training and test sets
cancer = load_breast_cancer()
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# logreg using all features
# print precision and recall
# and plot confusion matrix(using test set)
problemI(LogisticRegression(random_state=0,max_iter=4000))
```

```python
#Create 30 LogReg models by incrementally changing the number of features.
problemII(LogisticRegression(random_state=0, max_iter=4000))
```

```python
# svm linear kernel using all features
# print precision and recall
# and plot confusion matrix(using test set)
problemI(SVC(kernel='linear'))
```

```python
#Create 30 smv linear kernel models by incrementally changing the number of features.
# (a) Recall vs #faetures (b) Precision vs #features curves.
problemII(SVC(kernel='linear'))
```

```python
# svm gaussian kernel using all features
# print precision and recall
# and plot confusion matrix(using test set)
problemI(SVC(kernel='rbf'))
```
